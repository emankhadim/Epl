```python
import pandas as pd
import numpy as np
```

```python
path =("Desktop\epl2020.csv")
df = pd.read_csv(path)
df.head()
```

```python
df.shape
```

```python
#drop the column unnamed because this column is in no use
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df= df.reset_index(drop=True)
df.head()
```

# 1) What are the teams Performances in Home and Away?

```python
teams = df.teamId.unique()
teams
```

```python
mask=df['h_a']=='h'
df1=df.loc[mask]
df2=df1[['teamId','pts']]
df2=df2.groupby(['teamId']).agg({'pts':np.sum})
df2
```

```python
mask1=df['h_a']=='a'
df1=df.loc[mask1]
df3=df1[['teamId','pts']]
df3=df3.groupby(['teamId']).agg({'pts':np.sum})
df3.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
width = 0.3
ypos1=np.arange(len(df2))
ypos2=np.arange(len(df3))
plt.yticks(ypos1,df2.index)
plt.yticks(ypos2,df3.index)
plt.barh(ypos1,df2.pts,label="Home Wins")
plt.barh(ypos2,df3.pts,label="Away Wins")
plt.xlabel("Teams")
plt.ylabel("Points")
plt.title("Home and Away Wins")
plt.legend()
plt.show()

#stack bar plot
```

# 3) Goalkeepers Ranking Based on their Performance

```python
path1 =("Desktop\players.csv")
df7 = pd.read_csv(path1)
df7.head()
```

```python
players = df7.full.unique()
player_goals = {}
for player in players:
    mins = df7[df7['full'] == player].minutes.sum()/90.
    if mins >= 10:
        player_goals[player] = df7[df7['full'] == player].goals_scored.sum()/mins
#player_goals

top_players_goals = sorted(player_goals.keys(), key=player_goals.get, reverse=True)[:20]
top_players_goals
```

# 4)Comparing Expected performance by teams(made by supercomputers before event) vs Actual performance

```python
# Get the actual and expected points for each team
actual_pts = {}
expected_pts = {}
for team in teams:
  actual_pts[team] = df[df['teamId'] == team].pts.sum()
  expected_pts[team] = df[df['teamId'] == team].xpts.sum()
#pts_difference= {team: actual_pts[team] - expected_pts[team] for team in teams}
#orted_teams = sorted(pts_difference.keys(), key=pts_difference.get)
#pts_difference
```

```python
ypos1=np.arange(len(actual_pts))
pos = [i+0.45 for i in ypos1]
ypos2=np.arange(len(expected_pts))
plt.yticks(ypos1,actual_pts.keys())
plt.yticks(ypos2,expected_pts.keys())
plt.barh(ypos1,actual_pts.values(),0.45,label="Actual Points")
plt.barh(pos,expected_pts.values(),0.45,label="Expected Points")
plt.xlabel("Teams")
plt.ylabel("Points")
plt.title("Actual vs Expected Points")
plt.legend()
plt.show()
```
